```python
from sympy import  *
import numpy as np
import math
import cmath
from sympy.abc import s
from IPython.display import display, Math
from splane import pzmap, bodePlot
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
from scipy import signal
from sympy import symbols, solve

#Filtro Pasa Todo, punto 1)A)
num=[1, -1]
den=[1, 1]
my_tf = TransferFunction(num, den)
print(my_tf)
w, mag, phase = signal.bode(my_tf)
plt.figure()
plt.semilogx(w, mag)    # Bode magnitude plot
plt.figure()
plt.semilogx(w, phase) 
pzmap(my_tf)

#Punto 2
#A)
num=[0.25, 0, 1]
den=[1, 1.41, 1]
my_tf = TransferFunction(num, den)
print(my_tf)
w, mag, phase = signal.bode(my_tf)
plt.figure()
plt.semilogx(w, mag)    # Bode magnitude plot
plt.figure()
plt.semilogx(w, phase) 
pzmap(my_tf)

#B)
num=[1, 0.707, 1]
den=[1, 1.41, 1]
my_tf = TransferFunction(num, den)
print(my_tf)
w, mag, phase = signal.bode(my_tf)
plt.figure()
plt.semilogx(w, mag)    # Bode magnitude plot
plt.figure()
plt.semilogx(w, phase) 
pzmap(my_tf)

#C)
num=[1, -1.41*0.5, 0.25]
den=[1, 1.41*0.5, 0.25]
my_tf = TransferFunction(num, den)
print(my_tf)
w, mag, phase = signal.bode(my_tf)
plt.figure()
plt.semilogx(w, mag)    # Bode magnitude plot
plt.figure()
plt.semilogx(w, phase) 
pzmap(my_tf)

#D)
num=[1, 0, 16]
den=[1, 1.41, 1]
my_tf = TransferFunction(num, den)
print(my_tf)
w, mag, phase = signal.bode(my_tf)
plt.figure()
plt.semilogx(w, mag)    # Bode magnitude plot
plt.figure()
plt.semilogx(w, phase) 
pzmap(my_tf)
```
